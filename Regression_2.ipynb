Q3. When is it more appropriate to use adjusted R-squared?

Adjusted R-squared is a modified version of the R-squared metric that adjusts for the number of predictors in the model. It provides a more accurate measure of the goodness of fit for regression models, particularly when dealing with multiple predictors.

Situations to Use Adjusted R-squared:

- When you have more than one independent variable, adjusted R-squared is more appropriate as it accounts for the number of predictors. Unlike regular R-squared, which can artificially increase with the addition of more variables, adjusted R-squared penalizes the model for including unnecessary predictors.
- When comparing the performance of different regression models, especially those with a different number of predictors, adjusted R-squared provides a fair comparison by adjusting for the number of predictors. This helps to identify whether the additional predictors are actually contributing to the model or just adding noise.
- During the model selection process, adjusted R-squared can help in selecting the optimal model by balancing model complexity and explanatory power. It discourages the inclusion of irrelevant variables that do not improve the model significantly.
- Adjusted R-squared helps to mitigate overfitting by reducing the tendency to include too many predictors in the model. Overfitting occurs when the model captures noise in the training data, which does not generalize to new data.

Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics
calculated, and what do they represent?

## Mean Squared Error (MSE)
It measures the square root of the average discrepancies between a datasets actual values and projected values. MSE is frequently utilized in regression issues and is used to assess how well predictive models work. For a dataset containing n data points, the MSE calculation formula is:

MSE = \frac{1}{n} \sum ^{n} _{i=1} (x_{i} - y_{i})^2      

where:

xi represents the actual or observed value for the i-th data point.

yi represents the predicted value for the i-th data point.

## Mean Absolute Error (MAE)
Its a measurement of the typical absolute discrepancies between a datasets actual values and projected values. The formula to calculate MAE for a data with n data points is:

MAE =  \frac{1}{n} \sum ^{n} _{i=1} |x _{i} - y _{i}|      

Where:

xi represents the actual or observed values for the i-th data point.

yi represents the predicted value for the i-th data point.

## Root Mean Squared Error (RMSE)
It is a usually used metric in regression analysis and machine learning to measure the accuracy or goodness of fit of a predictive model, especially when the predictions are continuous numerical values. The RMSE quantifies how well the predicted values from a model align with the actual observed values in the dataset.

To obtain the RMSE, simply take the square root of the MSE. The formula for RMSE for a data with n data points is as follows:

RMSE = \sqrt{\frac{1}{n}\sum ^{n} _{i=1}(x_{i} - y_{i})^2 }

Q5. Discuss the advantages and disadvantages of using RMSE, MSE, and MAE as evaluation metrics in
regression analysis.

1. Mean Squared Error (MSE)

Advantages:

- Penalizes Larger Errors: Squaring the errors gives more weight to larger errors, which can be useful if larger errors are particularly undesirable.
- Mathematically Convenient: Differentiable, making it easier to compute gradients for optimization algorithms.

Disadvantages:

- Sensitive to Outliers: Because it squares the errors, MSE can be heavily influenced by outliers.
- Not Intuitive: The units of MSE are the square of the units of the dependent variable, making it less interpretable.

2. Root Mean Squared Error (RMSE)

Advantages:

- Intuitive Interpretation: The units of RMSE are the same as the units of the dependent variable, making it more interpretable.
- Sensitive to Large Errors: Like MSE, it penalizes larger errors, but the square root reduces the impact slightly.

Disadvantages:

- Still Sensitive to Outliers: While slightly less sensitive than MSE, RMSE can still be significantly influenced by outliers.

3. Mean Absolute Error (MAE)

Advantages:

- Robust to Outliers: Less sensitive to outliers compared to MSE and RMSE because it does not square the errors.
- Intuitive Interpretation: The units of MAE are the same as the units of the dependent variable, making it easy to interpret.

Disadvantages:

- Less Sensitive to Large Errors: Does not penalize large errors as heavily as MSE and RMSE, which may be a disadvantage if large errors are particularly problematic.
- Not Differentiable Everywhere: The absolute value function is not differentiable at zero, which can complicate the use of gradient-based optimization methods.

Q7. How do regularized linear models help to prevent overfitting in machine learning? Provide an
example to illustrate.

Regularized linear models, such as Ridge regression and Lasso regression, help prevent overfitting in machine learning by adding a penalty term to the cost function that limits the magnitude of the model coefficients. This penalty term discourages the model from assigning too much importance to any one feature, thereby reducing the model's sensitivity to noise and improving its generalization performance.

For example, let's consider a regression problem where we want to predict the sale price of houses based on a set of features, such as the number of bedrooms, the size of the lot, and the location of the house. We have a training set of 1000 houses, and we want to build a model that can accurately predict the sale price of new houses.
Without regularization, we might train a linear regression model that fits the training set very closely, with a high coefficient assigned to each feature. However, this model is likely to overfit to the training set, meaning it will perform poorly on new data that it has not seen before.

To prevent overfitting, we can add a penalty term to the cost function that limits the magnitude of the model coefficients. For example, we can use Ridge regression or Lasso regression to add an L2 or L1 penalty term, respectively, to the cost function.

With regularization, the model is encouraged to assign lower coefficients to less important features, which helps to prevent overfitting and improve its generalization performance. For example, in Lasso regression, the L1 penalty term can force some coefficients to be exactly zero, effectively removing those features from the model.

In summary, regularized linear models help prevent overfitting by adding a penalty term to the cost function that limits the magnitude of the model coefficients. By encouraging the model to assign lower coefficients to less important features, these models can improve the generalization performance of the model and reduce its sensitivity to noise.



Q8. Discuss the limitations of regularized linear models and explain why they may not always be the best
choice for regression analysis.

- Regularization techniques (like Lasso and Ridge) can produce models that are harder to interpret, especially when they shrink coefficients towards zero or entirely to zero (as in Lasso). This can obscure the understanding of the underlying relationship between predictors and the outcome.
- Regularized models require that features are standardized or normalized, as the penalty terms are sensitive to the scale of the data. This adds an extra preprocessing step and can be problematic if the dataset has features on very different scales.
- Regularized models may not perform well if the data quality is poor or if the relevant features are not included in the model. They assume that the true relationship can be captured by penalizing the size of coefficients, which may not be true in all cases.
- While regularization reduces variance by preventing overfitting, it can also introduce bias into the model. This bias-variance trade-off means that the model may underfit the data if the regularization is too strong.
- Regularized models introduce additional hyperparameters (like the regularization strength) that need to be carefully tuned. This tuning process can be computationally intensive and complex.
- While Ridge regression can help mitigate multicollinearity by shrinking coefficients, Lasso can struggle when predictors are highly correlated, arbitrarily selecting one and shrinking others to zero.
