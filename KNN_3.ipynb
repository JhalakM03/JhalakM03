**Q3. Write a Python code snippet to find the optimal value of K for the KNN classifier algorithm using cross-validation on load_iris dataset in sklearn.datasets.**

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
x,y=load_iris(return_X_y=True)
```

```python
k_values=range(1,31)
cv_scores=[]
```

```python
for k in k_values:
  knr=KNeighborsClassifier(n_neighbors=k)
  scores=cross_val_score(knr,x,y,scoring='accuracy',cv=10)
  cv_scores.append(scores.mean())
```

```python
optimal_k = k_values[np.argmax(cv_scores)]
print(f'The optimal value of k is {optimal_k}')
```

**Q6. Implement a function to standardise the features before applying KNN classifier.**

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.metrics import classification_report
```

```python
def standardize(x,y,n=5, test_size=0.33,cv=5):
  xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=42)
  scaler=StandardScaler()
  xtrain=scaler.fit_transform(xtrain)
  xtest=scaler.transform(xtest)
  knnc=KNeighborsClassifier(n_neighbors=n)
  knnc.fit(xtrain,ytrain)
  scores=cross_val_score(knnc,xtrain,ytrain,cv=cv,scoring='accuracy')
  cs_score_mean=np.mean(scores)
  ypred=knnc.predict(xtest)
  class_report=classification_report(ytest,ypred)
  print('Classification Report on Test data :\n')
  print(classification_report(ytest, ypred))
  return xtrain,xtest,ytrain,ytest,knnc,cs_score_mean
```

```python
X,y=load_breast_cancer(return_X_y=True)
```

```python
xtrain, xtest, ytrain, ytest, knnc, cs_score_mean = standardize(X,y,n=5,test_size=0.33,cv=5)
```

**Q7. Write a Python function to calculate the euclidean distance between two points.**

```python
import math

def euclidean_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Both points must have the same number of dimensions")

    squared_diffs = [(p1 - p2) ** 2 for p1, p2 in zip(point1, point2)]

    return math.sqrt(sum(squared_diffs))

point_a = [3, 4]
point_b = [6, 8]
distance = euclidean_distance(point_a, point_b)
print(f"The Euclidean distance between the points is: {distance}")
```

**Q8. Write a Python function to calculate the manhattan distance between two points.**

```python
def manhattan_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Both points must have the same number of dimensions")

    absolute_diffs = [abs(p1 - p2) for p1, p2 in zip(point1, point2)]
    return sum(absolute_diffs)

point_a = [3, 4]
point_b = [6, 8]
distance = manhattan_distance(point_a, point_b)
print(f"The Manhattan distance between the points is: {distance}")
```
