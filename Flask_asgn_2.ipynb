Q1. Explain GET and POST methods.

GET: The GET method is used to request data from a specified resource.It is considered less secure for sensitive data since parameters are visible in the URL. It is used for requests where data is being retrieved from the server. For example, fetching information about a product, displaying a page, or searching.

POST: The POST method is used to submit data to be processed to a specified resource.  It is considered more secure for sensitive data since parameters are not directly visible in the URL. It is used for requests where data is being submitted to the server. For example, submitting a form, creating a new resource, or updating a resource.

Q2. Why is request used in Flask?

The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods. It allows the Flask application to read and process data sent by a client in a request, such as form data or query parameters, and use that data to generate an appropriate response.

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect a user to a different URL. The redirect() function takes two arguments: the URL that the user should be redirected to, and the status code. The status code is a number that indicates the reason for the redirect. This can be useful for a number of reasons, such as:

1. To redirect a user to a new page after they have successfully logged in.
2. To redirect a user to a different page if they try to access a page that they are not authorized to view.
3. To redirect a user to a different page if they try to access a page that does not exist.
