Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError class is used general way to handle exceptions related to looking up or accessing itemsin data structures like lists or dictionaries.

```python
try:
    list_1 = [5,8,5,2]
    value = list_1[4]
except IndexError as e:
    print(f'IndexError: {e}')
```

```python
try:
    dict_2 = {'jan':'one', 'Feb':'two', 'Mar':'three'}
    value = dict_2['jun']
except KeyError as e:
    print(f'KeyError: {e}')
```

Q6. List down some best practices for exception handling in python.

1. Write the code in Try block which might raise an error and catch the exception using the except block.
2. Catch Specific exceptons when possible rather than using broader Exception class. This gives much better clarity while handling the error.
3. Clean up with Finally block because it will be executed whether the exception was raised or not.
4. Provide meaningful error messages to the user with print statement or by logging method.
