Flask is a web framework that allows developers to build lightweight web applications quickly and easily with Flask Libraries.
Advantages:
Flask is a lightweight backend framework with minimal dependencies.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask
app = Flask(__name__)

@app.route('/Hello12')
def Hello_world():
    return "<h3>Hello World!!</h3>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

App Routing means mapping the URLs to a specific function that will handle the logic for that URL. 
App Routing makes the web app simple to handle because each endpoint is handled by a distinct funtion.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask
app = Flask(__name__)

dict = {'Company Name': 'ABC Corporation',
'Location': 'India',
'Contact Detail': '999-999-9999'}

@app.route('/')
def details():
    return dict

@app.route('/welcome')
def welcome():
    return f"Welcome to {dict['Company Name']}"
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
