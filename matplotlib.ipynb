Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

The scatter() plots serve as a visual tool to explore and analyze the relationships between variables, utilizing dots to depict the connection between them. 

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
plt.figure(figsize=(4,3))
plt.scatter(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is scatter plot')
plt.grid()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots in a single figure. By sepecifing the no. of rows and columns.

```python
import numpy as np
#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2:
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4:
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Line1')
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Line2')
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Line 3')
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Line 4')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart is used to compare between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(4,4))
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid()
```

```python
plt.figure(figsize=(4,4))
plt.barh(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.grid()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. 

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
