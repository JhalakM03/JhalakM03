Q1. What is the mathematical formula for a linear SVM?

The mathematical formula for a linear Support Vector Machine (SVM) involves finding the optimal hyperplane that best separates two classes in a binary classification problem. The goal of a linear SVM is to maximize the margin between the two classes while minimizing classification errors.

- Hyperplane Equation: A hyperplane in an n-dimensional space can be defined as:

                w⋅x−b=0
                
where, w=(w1,w2,....wn) is the weight vector (normal to the hyperplane).
x=(x1,x2,....xn) is the input feature vector.
b is the bias term (or intercept).

Decision Function: The decision function for classifying a new data point x is given by:

                f(x)=w⋅x−b
                
A data point x is classified as:
                - Class 1 if f(x)>=0
                - Class -1 if f(x) < 0
                
Constraints: For a dataset with m  samples (xi,yi) where yi ∈ {-1,1} (representing the two classes), the constraints for correctly classifying all points are:

                yi(w.xi-b)>=1, ∀i=1,2,...m
                
This constraint ensures that all points are correctly classified and lie on the correct side of the margin.

Optimization Objective: The objective of the linear SVM is to maximize the margin, which is defined as 2/∥w∥. Maximizing the margin is equivalent to minimizing ∥w∥^2/2. Thus, the optimization problem becomes:

                min(w.b) 1/2 ∥w∥^2
                
Subject to: 
                
                yi(w.xi-b)>=1, ∀i1,2,...m

Q3. What is the kernel trick in SVM?

The kernel trick in Support Vector Machines (SVM) is a powerful technique that allows SVMs to perform classification in high-dimensional or even infinite-dimensional spaces without explicitly computing the coordinates of the data in that space. This is particularly useful for dealing with non-linear data that cannot be separated by a straight line (hyperplane) in its original feature space.

The kernel trick allows SVMs to learn non-linear decision boundaries by mapping the data into higher-dimensional spaces where a linear separator can be found. Instead of computing the transformation ϕ(x) for every data point explicitly, the kernel trick allows SVMs to compute the inner products directly in the original space, which is computationally efficient.Different kernel functions can be chosen based on the nature of the data and the problem, providing flexibility in finding the optimal decision boundary. 

Different types of kernel functions can be used, such as polynomial, radial basis function (RBF), and sigmoid kernels.
The most commonly used kernel function is the RBF kernel, which is defined as:

            K(x, y) = exp(-γ * ||x - y||^2)  
            
γ is a parameter that defines the spread of the kernel.

The RBF kernel has a high value when the feature vectors are close to each other and decreases exponentially as the distance between them increases.

By using the kernel trick, the SVM can find a non-linear decision boundary in the higher-dimensional space, without explicitly computing the mapping. This allows the SVM to handle complex, non-linearly separable data efficiently.
In summary, the kernel trick in SVM is a technique to implicitly map input data into a higher-dimensional space using a kernel function, which allows the SVM to handle non-linearly separable data and find a non-linear decision boundary in the higher-dimensional space.

Q4. What is the role of support vectors in SVM Explain with example.

Support Vectors play a crucial role in the functioning of a Support Vector Machine (SVM). They are the data points that lie closest to the decision boundary (hyperplane) and are the most influential in determining the position and orientation of the hyperplane.

In SVM, the goal is to find the hyperplane that best separates the two classes in the feature space with the maximum margin. The margin is the distance between the hyperplane and the nearest data points from either class. Support vectors are the data points that are exactly on the edge of this margin. These points are crucial because they are the closest to the decision boundary and effectively "support" the hyperplane.

The SVM algorithm aims to maximize the margin, which is the distance between the hyperplane and the support vectors. By maximizing this margin, SVM ensures that the classifier is robust and has a lower generalization error, meaning it performs better on unseen data. The position and orientation of the hyperplane are solely determined by the support vectors. Other data points that are further away from the margin do not affect the hyperplane. This makes SVM particularly efficient because it only relies on a subset of data points (the support vectors) to construct the model. Support vectors help make SVM robust to outliers.



In this example, the two data points closest to the decision boundary are the support vectors. These are the points that "support" the decision boundary and determine its position and orientation. All other data points have no effect on the decision boundary.
The margin of the SVM is the distance between the decision boundary and the closest data points of each class. In this example, the margin is shown as the two dashed lines. The SVM algorithm aims to maximize this margin, while still correctly classifying all data points.



Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

1. Hyperplane
In SVM, a hyperplane is a decision boundary that separates the data into different classes. In a binary classification problem, a hyperplane is a line that separates the data points of one class from the data points of the other class. In a multiclass classification problem, a hyperplane is a plane or a higher-dimensional hyperplane that separates the data points of each class from the data points of the other classes.



In this example, the blue and red dots represent two different classes of data points. The hyperplane is a line that separates the blue dots from the red dots. Any new data point that falls on the blue side of the line will be classified as belonging to the blue class, while any new data point that falls on the red side of the line will be classified as belonging to the red class.

2. Marginal Plane
The marginal plane is the hyperplane that is equidistant from the support vectors on each side. The distance between the marginal plane and the closest support vector is called the margin, and it is maximized in SVM.

3. Hard Margin
In SVM, the hard margin method is used when the data is linearly separable, meaning that a hyperplane can be found that perfectly separates the data into two classes. The hard margin method tries to find the hyperplane that maximizes the margin while ensuring that all data points are correctly classified.

4. Soft Margin
In SVM, the soft margin method is used when the data is not linearly separable, meaning that a hyperplane cannot be found that perfectly separates the data into two classes. The soft margin method allows for some misclassification of the data points in order to find a hyperplane that separates the data as well as possible.

## Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl

```python
from sklearn.datasets import load_iris
df=load_iris()
```

```python
print(df.DESCR)
```

```python
df.keys()
```

```python
import pandas as pd
X=pd.DataFrame(df.data, columns=df.feature_names)
X.head()
```

```python
df.target_names
```

```python
y=pd.DataFrame(df.target, columns=['class'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

##  Train a linear SVM classifier on the training set and predict the labels for the testing set

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',C=1.0)
svm.fit(X_train,y_train.values.flatten())
```

```python
y_pred=svm.predict(X_test)
```

```python
y_pred
```

## Compute the accuracy of the model on the testing setl

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy=accuracy_score(y_test,y_pred)
print('Testing accuracy: ',accuracy)
```

```python
import seaborn as sns
cf=confusion_matrix(y_test,y_pred)
sns.heatmap(cf,annot=True)
```

## Plot the decision boundaries of the trained model using two of the featuresl

```python
import matplotlib.pyplot as plt
df_train=pd.concat([X_train,y_train],axis=1)
sns.scatterplot(data=df_train, x='sepal length (cm)', y='sepal width (cm)', hue='class')
plt.title('Sepal Length vs Sepal width hueplot')
plt.show()
```

```python
sns.scatterplot(data=df_train, x='petal length (cm)', y='petal width (cm)', hue='class')
plt.title('petal Length vs petal width hueplot')
plt.show()
```

```python
sns.pairplot(data=df_train,hue='class')
plt.show()
```

## Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import numpy as np
C=np.linspace(0.01,10,20)
```

```python
C
```

```python
from sklearn.metrics import classification_report
acc=[]
for i in C:
    model=SVC(kernel='linear',C=i)
    model.fit(X_train,y_train.values.flatten())
    y_test_pred=model.predict(X_test)
    acc.append(accuracy_score(y_test,y_test_pred))
    print(f'C Value: {i} \n')
    print(classification_report(y_test,y_test_pred))
    print('\n-----------------------------------------------------------------\n')
```

```python
import matplotlib.pyplot as plt
plt.plot(C, acc)
plt.xlabel('C value')
plt.ylabel('Testing Accuracy')
plt.title('Accuracies for different values of C')
plt.show()
```
